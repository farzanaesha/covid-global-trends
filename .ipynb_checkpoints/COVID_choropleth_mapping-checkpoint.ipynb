## 📦 Importing required libraries

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objs as go
from plotly.offline import iplot
```

## 📂 Loading COVID-19 dataset


```python
df = pd.read_csv('epidemiology.csv')
df.head()
```

## 🧹 Cleaning COVID-19 dataset

```python
df_cleaned=df.dropna()
```

## Describe Data

```python
df.info()
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'])
oldest_date = df['date'].min().date()
latest_date = df['date'].max().date()

print("📉 Oldest date:", oldest_date)
print("📈 Latest date:", latest_date)
```

## 🧮 Calculate death and survival rates

```python
# Step 1: Filter out rows with zero confirmed cases
df_clean = df[df['new_confirmed'] > 0].copy()

# Step 2: Calculate death and survival rates
df_clean['Death Rate'] = df_clean['new_deceased'] / df_clean['new_confirmed']
df_clean['Survival Rate'] = df_clean['new_recovered'] / df_clean['new_confirmed']

# Step 3: Cumulative totals from full dataset (not filtered)
total_deaths = df['new_deceased'].sum()
total_recovered = df['new_recovered'].sum()

# Step 4: Print results
print("📊 Cumulative Deaths:", total_deaths)
print("📊 Cumulative Recovered:", total_recovered)
print("📉 Average Death Rate:", round(df_clean['Death Rate'].mean() * 100, 2), "%")
print("📈 Average Survival Rate:", round(df_clean['Survival Rate'].mean() * 100, 2), "%")
```

## 🌍Mapping COVID-19 Mortality by Region

```python
# Group by location_key and sum daily deaths
total_deaths_by_location = df.groupby('location_key')['new_deceased'].sum()
# Filter locations with total deaths > 0
filtered_deaths = total_deaths_by_location[total_deaths_by_location > 0]

# Convert Series to DataFrame
total_deaths_df = total_deaths_by_location.reset_index()

# Rename columns for clarity (optional)
total_deaths_df.columns = ['location_key', 'total_deaths']

# Display result
#print(total_deaths_df)
```

```python
# 🌍 Prepare choropleth data (example: death rate)
chorodata = go.Choropleth(
    locations=total_deaths_df['location_key'],
    locationmode='ISO-3',
    z=total_deaths_df['total_deaths'],
    colorscale='Reds',
    colorbar_title='Total Deaths',
    text=total_deaths_df['total_deaths']
)

# 🗺️ Define layout
layout = go.Layout(
    title='Global COVID-19 Mortality and Survival Patterns: A Choropleth Visualization',
    width=1000,
    height=600,
    geo=dict(
        showframe=False,
        projection=dict(type='natural earth')
    )
)

# 🖼️ Create and display the figure
fig = go.Figure(data=[chorodata], layout=layout)
iplot(fig)
```

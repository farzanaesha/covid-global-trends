```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import iplot
import cufflinks as cf
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot
```

```python
# The official URL for the Our World in Data COVID-19 CSV 
#URL = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
covid= pd.read_csv('covid_analysis.csv')
```

```python
#Converts the 'date' column to the proper datetime format
covid['date'] = pd.to_datetime(covid['date'])
```

```python
#Removes aggregate rows like "World," "High-income countries,","Asia," etc. leaving only individual country data.
covid_countries = covid.dropna(subset=['continent']).copy()
```

```python
covid_countries.isnull().sum()
```

```python
# Calculates Total Deaths per 1 million people (Standardized Mortality Rate)
covid_countries['deaths_per_million'] = (covid_countries['total_deaths'] / covid_countries['population']) * 1_000_000

# Calculates Case Fatality Rate (CFR) 
covid_countries['case_fatality_rate'] = (covid_countries['total_deaths'] / covid_countries['total_cases']) * 100

covid_countries[['location', 'date', 'total_deaths', 'population', 'deaths_per_million','case_fatality_rate']].sample(5)
```

```python
covid_valid_mortality = covid_countries[
    covid_countries['total_deaths'].notnull() & 
    (covid_countries['total_deaths'] > 0)
].copy()
covid_notnull_mortality = covid_countries[
    covid_countries['total_deaths'].notnull()].copy()

# Sort by location and then by date descending
covid_latest_valid_mortality =covid_valid_mortality.sort_values(by=['location', 'date'], ascending=[True, False])\
               .drop_duplicates(subset=['location'], keep='first')
covid_latest_notnull_mortality =covid_notnull_mortality.sort_values(by=['location', 'date'], ascending=[True, False])\
               .drop_duplicates(subset=['location'], keep='first')

# Keep only the first row for each location (which is the latest date)
#covid_comparison = covid_latest

print(f"Original daily data rows: {len(covid_countries)}")
print(f"Valid comparison table rows (countries): {len(covid_latest_valid_mortality)}")
print(f"Notnull comparison table rows (countries): {len(covid_latest_notnull_mortality)}")
```

```python
print("📈 Top 10 Countries by Deaths Per Million (Total Impact):")
covid_latest_valid_mortality[['location', 'deaths_per_million', 'population']] \
    .sort_values(by='deaths_per_million', ascending=False) \
    .head(10)
    
```

```python
print("\n📉 Bottom 10 Countries by Deaths Per Million:")
# Filter out countries with zero deaths/cases to get meaningful 'bottom'
covid_latest_notnull_mortality[['location', 'deaths_per_million', 'population']] \
    .sort_values(by='deaths_per_million', ascending=True) \
    .head(10)
```

```python
#Visualize the smoothed daily deaths for a single country

country_name = 'Bangladesh' 
country_df = covid_countries[covid_countries['location'] == country_name]

fig = px.line(
    country_df, 
    x='date', 
    y='new_deaths_smoothed', 
    title=f'7-Day Smoothed Daily COVID-19 Deaths in {country_name}'
)
fig.show()
```

```python
from geopy.geocoders import Nominatim
import time

geolocator = Nominatim(user_agent="geoapi")

def get_coords(country):
    try:
        location = geolocator.geocode(country)
        time.sleep(1)  # avoid rate limit
        return pd.Series([location.latitude, location.longitude])
    except:
        return pd.Series([None, None])

covid_latest_valid_mortality[['latitude', 'longitude']] = covid_latest_valid_mortality['location'].apply(get_coords)
covid[['latitude', 'longitude']] = covid['location'].apply(get_coords)
```

```python
# Create choropleth
fig = px.choropleth(
    covid_latest_valid_mortality,
    locations='iso_code',
    color='deaths_per_million',
    hover_name='location',
    hover_data={
        'deaths_per_million': True,
        'total_deaths': True,
        'population': True,
        'iso_code': False
    },
    projection='natural earth',
    color_continuous_scale='Reds',
    title='Global COVID-19 Standardized Mortality Rate (Total Deaths per Million)',
    width=1000,
    height=700
)

# Prepare top 5 legend text
top5 = covid_latest_valid_mortality.sort_values(by='deaths_per_million', ascending=False).head(5)
legend_text = '<b>Top 5 Death Rates</b><br>'
for i, row in enumerate(top5.itertuples(), 1):
    legend_text += f"{i}. {row.location} ({row.deaths_per_million:.1f}/M)<br>"

# Add fixed annotation box (like a legend)
fig.add_annotation(
    text=legend_text,
    x=0, y=1,  # top-left corner
    xref='paper', yref='paper',
    align='left',
    showarrow=False,
    font=dict(size=12, color='black'),
    bordercolor=None,
    borderwidth=1,
    bgcolor='white',
    opacity=0.8
)

fig.update_geos(
    showframe=False,
    showcoastlines=False,
    showland=True
)

fig.update_layout(
    coloraxis_colorbar=dict(title='Deaths per Million')
)

fig.show()
```

```python
# The official URL for the Our World in Data COVID-19 CSV 
#URL = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
df= pd.read_csv('covid_analysis.csv')
# Load a CSV with ISO codes + lat/lon
geo_df = pd.read_csv('geo_code.csv')

# Merge with your COVID data
covid= df.merge(geo_df[['name','latitude', 'longitude']], left_on='location', right_on='name',how='left')
```

```python
#Converts the 'date' column to the proper datetime format
covid['date'] = pd.to_datetime(covid['date'])
```

```python
#Removes aggregate rows like "World," "High-income countries,","Asia," etc. leaving only individual country data.
covid_countries = covid.dropna(subset=['continent']).copy()
```

```python
covid_countries.isnull().sum()
```

```python
covid_valid_mortality = covid_countries[
    covid_countries['total_deaths'].notnull() & 
    (covid_countries['total_deaths'] > 0)
].copy()
covid_notnull_mortality = covid_countries[
    covid_countries['total_deaths'].notnull()].copy()

# Sort by location and then by date descending
covid_latest_valid_mortality =covid_valid_mortality.sort_values(by=['location', 'date'], ascending=[True, False])\
               .drop_duplicates(subset=['location'], keep='first')
covid_latest_notnull_mortality =covid_notnull_mortality.sort_values(by=['location', 'date'], ascending=[True, False])\
               .drop_duplicates(subset=['location'], keep='first')

# Keep only the first row for each location (which is the latest date)
#covid_comparison = covid_latest

print(f"Original daily data rows: {len(covid_countries)}")
print(f"Valid comparison table rows (countries): {len(covid_latest_valid_mortality)}")
print(f"Notnull comparison table rows (countries): {len(covid_latest_notnull_mortality)}")
```

```python
# 2. Find the MAXIMUM stringency index for each country over the entire period
# The 'stringency_index' column is what we are interested in.
stringency_data = covid_countries.groupby('location')['stringency_index'].max().reset_index()

# Rename columns for clarity in the final map
stringency_data.columns = ['Country', 'Max_Stringency_Index']

# 3. Handle potential NaN values (countries with no data)
stringency_data = stringency_data.dropna(subset=['Max_Stringency_Index'])
```
